```python
import re
import matplotlib.pyplot as plt
import numpy as np

# File path to your log file
log_file_path = 'logs.txt'

# Regular expressions to extract values
epoch_pattern = r"Epoch (\d+), Train Loss: ([\d\.]+), Val Loss: ([\d\.]+), Accuracy: ([\d\.]+)"

# Lists to store extracted values
epochs = []
train_losses = []
val_losses = []
accuracies = []

# Read the log file
with open(log_file_path, 'r') as file:
    log_data = file.readlines()

# Extract training and validation data
for line in log_data:
    epoch_match = re.search(epoch_pattern, line)

    if epoch_match:
        epochs.append(int(epoch_match.group(1)))
        train_losses.append(float(epoch_match.group(2)))
        val_losses.append(float(epoch_match.group(3)))
        accuracies.append(float(epoch_match.group(4)))

# Plotting
plt.figure(figsize=(8, 12))

# 1️⃣ Training & Validation Loss
plt.subplot(2, 1, 1)
plt.plot(epochs, train_losses, marker='o', label='Training Loss', color='blue')
plt.plot(epochs, val_losses, marker='o', label='Validation Loss', color='red', linestyle='dashed')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training & Validation Loss')
plt.legend()
plt.grid(True)

# 2️⃣ Accuracy
plt.subplot(2, 1, 2)
plt.plot(epochs, accuracies, marker='o', label='Accuracy', color='green')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
