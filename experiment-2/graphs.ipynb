```python
import re
import matplotlib.pyplot as plt

# File path to your log file
log_file_path = 'without_hidden_logs.txt'

# Regular expression to extract epochs and training loss
no_hidden_layer_pattern = r"Epoch (\d+), Training Loss: ([\d\.]+)$"

# Lists to store extracted values
epochs_no_hidden = []
train_losses_no_hidden = []

# Read the log file
with open(log_file_path, 'r') as file:
    log_data = file.readlines()

# Extract training loss data
for line in log_data:
    match = re.search(no_hidden_layer_pattern, line)
    if match:
        epochs_no_hidden.append(int(match.group(1)))
        train_losses_no_hidden.append(float(match.group(2)))

# Plot Training Loss
plt.figure(figsize=(8, 5))
plt.plot(epochs_no_hidden, train_losses_no_hidden, marker='o', label='Training Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('Training Loss (No Hidden Layer)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

# File path to your log file
log_file_path = 'logs.txt'

# Regular expression to extract epoch, training loss, and accuracy
pattern = r"Epoch (\d+), Training Loss: ([\d\.]+), Accuracy: ([\d\.]+)"

# Dictionaries to store values (ensures latest value for each epoch)
loss_dict = {}
accuracy_dict = {}

# Read the log file
with open(log_file_path, 'r') as file:
    log_data = file.readlines()

# Extract data
for line in log_data:
    match = re.search(pattern, line)
    if match:
        epoch = int(match.group(1))
        loss = float(match.group(2))
        accuracy = float(match.group(3))
        
        loss_dict[epoch] = loss  # Store the latest loss for each epoch
        accuracy_dict[epoch] = accuracy  # Store the latest accuracy for each epoch

# Sort data by epochs
epochs_sorted = sorted(loss_dict.keys())
losses_sorted = [loss_dict[epoch] for epoch in epochs_sorted]
accuracies_sorted = [accuracy_dict[epoch] for epoch in epochs_sorted]

# Plot Training Loss
plt.figure(figsize=(8, 5))
plt.plot(epochs_sorted, losses_sorted, marker='o', linestyle='-', linewidth=2, markersize=6, color='blue', label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss (Hidden Layer)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Training Accuracy
plt.figure(figsize=(8, 5))
plt.plot(epochs_sorted, accuracies_sorted, marker='o', linestyle='-', linewidth=2, markersize=6, color='green', label='Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training Accuracy (Hidden Layer)')
plt.legend()
plt.grid(True)
plt.show()
```
